```python
from random import randint
class bank_account:
 
     
    def __init__(self):
        self.number = []
        self.pin = []
        self.name = []
        self.dob = []
        self.balance = []
        self.iin = 400000
   
    def main_screen(self):
            print(
                "________KIGGY BANK_______\n"
                "1. Create an account\n"
                "2. Deposit, Withdraw, Transfer, Account Details\n"
                "3. Exit"
            )
            main_menu_selection = str(input())
            if main_menu_selection == "1":
                self.create_acc()
            if main_menu_selection == "2":
                self.account_login()
            if main_menu_selection == "0":
                print("\n")
                print("Thank You, have a wonderful day!!")
    def account_login(self):
        print("\n")
        print("Enter your Account number:")
        entered_acc_number = int(input())
    
        print("Enter your PIN:")
        entered_pin_number = int(input())

        if str(entered_acc_number) in self.number and str(entered_pin_number) in self.pin:
            print("\n")
            print("You have logged in successfully!")
            self.account_services(entered_acc_number)
        else:
            print("\n")
            print("you have entered wrong PIN!")
            print("\n")
            self.main_screen()
            
    def account_services(self, entered_acc_number):
        print(
            "\n1. Account Details\n"
            "2. Deposit\n"
            "3. Withdraw\n"
            "4. Transfer\n"
            "5. Log out"
        )
        account_service_selection = str(input())
        for i in range(0,len(self.number)):
            if str(entered_acc_number) == self.number[i]:
                if account_service_selection == "1":
                    print("\n")
                    print("Account Holder's Name: ", self.name[i])
                    print("\n")
                    print("DOB: ", self.dob[i])
                    print("\n")
                    print("Account Balance: ", self.balance[i])
                    print("\n")
                    print("\n")
                    self.account_services(entered_acc_number)
    
                elif account_service_selection == "5":
                    print("\n")
                    print("You have logged out successfully!")
                    print("\n")
                    self.main_screen()
                elif account_service_selection == "4":
                    self.transfer(entered_acc_number)
                elif account_service_selection == "3":
                    self.withdraw(entered_acc_number)
                elif account_service_selection == "2":
                    self.deposit(entered_acc_number)
            
            
    def transfer(self, myacc):
        toacc = (input("Enter account number for whom to transfer money "))
        money = int(input("Enter amount to transfer $ "))
        for i in range(0,len(self.number)):
            if toacc == self.number[i]:
                self.balance[i] = int(self.balance[i]) + money
            if str(myacc) == self.number[i]:
                self.balance[i] = int(self.balance[i]) - money
                print("Updated Balance : ", self.balance[i])
        self.account_services(myacc)
        
 
    def withdraw(self, myacc):
        money = int(input("Enter amount to withdraw: "))
        for i in range(0,len(self.number)):
            if str(myacc) == self.number[i]:
                if int(self.balance[i]) >= money: 
                    self.balance[i] = int(self.balance[i]) - money
                    inp = int(input("Do you want to print a receipt?\n\t1. Yes\n\t2. No\n\t\t\t"))
                    if inp == 1:
                        print("Please collect your receipt from your nearest branch :) ")
                        print("Updated Balance : $", self.balance[i])
                        self.account_services(myacc)
                    elif inp == 2:
                        print("Updated Balance : $", self.balance[i])
                        self.account_services(myacc)
                else:
                    print("Insufficient Balance!!!")
                    inp = int(input("wanna try again?\n\t1. Yes\n\t2. No\n\t\t\t"))
                    if inp == 1:
                        self.withdraw(myacc)
                    elif inp == 2:
                        self.account_services(myacc)
                        
                        
    def deposit(self, myacc):
        
        money = int(input("Enter amount to deposit: "))
        for i in range(0,len(self.number)):
            if str(myacc) == self.number[i]: 
                    self.balance[i] = int(self.balance[i]) + money
                    inp = int(input("Do you want account receipt?\n\t1. Yes\n\t2. No\n\t\t\t"))
                    if inp == 1:
                        print("Please get your account receipt from the nearest Kiggy Bank :) ")
                        print("Updated Balance : $", self.balance[i])
                        self.account_services(myacc)
                    elif inp == 2:
                        print("Updated Balance : $", self.balance[i])
                        self.account_services(myacc)

        
         
    def create_acc(self):
        acc_number = "401020" + format(randint(0000000000, 9999999999), '010d')
        pin_number = format(randint(0000, 9999), '04d')
        
        self.number.append(acc_number)
        self.pin.append(pin_number)
        self.balance.append(25)
        
        print("\n")
        print("Welcome to Kiggy Bank")
        print("\n")
        nam = input("Please, enter your full name : ")
        self.name.append(nam)
        self.dob.append(input("Please enter your DOB in dd/mm/yyyy format : "))
        print("\n")
        print("Your account has been created")
        print("Your account number:")
        print(acc_number)
        print("Your PIN number:")
        print(pin_number)
        print("\n")
        print("Hello {0}, We are happy to have you as one of our customer.\nAs a welcome bonus, we will be giving you a $25 credit as an opening balance".format(nam))
        print("Please be aware of spam calls and spam messages. Call Kiggy Bank customer care  in case of emergency.")
        print("\n\n\n\t\t\t\t\t\t\t We are with you -- Kiggy Bank :)")
        print("\n")
        self.main_screen()
        
print(bank_account().main_screen())
    




```
